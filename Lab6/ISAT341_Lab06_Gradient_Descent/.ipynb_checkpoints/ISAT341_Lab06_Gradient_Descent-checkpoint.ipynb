## Examples with Simple Functions:

### Example 1: Demsonstrate Gradient Descent with a Univariant Function

Taking the function $J(\theta) = \theta^2 $ as an example:

#### Visualize the function

```python
# define the function f of x with the given equation
def J1(theta):
    return theta**2
```

The plot of this function is look like below

```python
theta = np.arange(-4,4,0.1)
cost = J1(theta)
plt.plot(theta,cost)
plt.xlabel("theta")
plt.ylabel("cost")
```

```python
def plot_gradient_descent(theta, cost, gradient):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.arange(theta-1,theta+1,0.1)
    y_vals = x_vals*gradient +(cost-gradient*theta) 
    color = next(axes._get_lines.prop_cycler)['color'] 
    plt.plot(x_vals, y_vals, color=color, linestyle='--')
    plt.scatter(theta,cost,color=color,s=50)
```

#### Implement Gradient Descent for the function

The gradient (i.e.,derivitaives) is:
$$\frac{d}{dx}J(\theta)= 2\theta$$  

Repeat to update parameter(s) until converge
$$ \theta_ùëó:= \theta_ùëó‚àí\alpha \frac{d}{dx}J(\theta)=\theta_ùëó‚àí\alpha*2\theta$$

```python
# gradient descent
# initialization
# inital the parameter(s)
theta_gd =3
# setup step size or learning rate
alpha = 0.4

# visualize the process of gradient descent
plt.plot(theta,cost)
plt.xlabel("theta")
plt.ylabel("cost")

# gradient descent algorithm: interatively taking steps in the steepest direction
for i in range(8):
    # display the progress and plot
    print("theta: % f, J(theta): % f" %(theta_gd, J1(theta_gd)))
    plot_gradient_descent(theta_gd,J1(theta_gd),2*theta_gd)
    # find new parameter and weight
    theta_gd = theta_gd - alpha * 2*theta_gd
print("theta: % f, J(theta): % f" %(theta_gd, J1(theta_gd)))
```

A theta as 0.000008 can build the best hypothsis for which the cost function is $J(\theta) = \theta^2$

### Example 2: Demsonstrate Gradient Descent with a Multivariant Function

Taking the function $J(\theta_0, \theta_1) = 2\theta_0^2+\theta_1^2 $ as an example:

#### Visualize the function

```python
theta_0 = np.arange(-10,10,0.1)
theta_1 = np.arange(-10,10,0.1)
```

```python
def J2(theta_0, theta_1):
    return 2*theta_0 ** 2 + theta_1 ** 2
```

```python
from mpl_toolkits import mplot3d
```

```python
X, Y = np.meshgrid(theta_0, theta_1)
Z = J2(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('theta_0')
ax.set_ylabel('theta_1')
ax.set_zlabel('cost');
ax.contour(X, Y, Z)
```

```python
cp = plt.contour(X,Y,Z)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('theta_0')
plt.ylabel('theta_1')
```

#### Implement Gradient Descent for the function

The gradients (i.e.,partial derivitaives) are:
$$\frac{\partial}{\partial \theta_0}J(\theta_0, \theta_1)=4\theta_0$$
$$\frac{\partial}{\partial \theta_1}J(\theta_0, \theta_1)=2\theta_1$$


Repeat to update parameter(s) until converge
$$ \theta_ùëó:= \theta_ùëó‚àí\alpha \frac{\partial}{\partial \theta_j}J(\theta)$$

```python
# gradient descent
# initialization
# inital the parameter(s)
theta_0_gd =-1
theta_1_gd =-2

# setup step size or learning rate
alpha = 0.4

# gradient descent algorithm: interatively taking steps in the steepest direction
for i in range(10):
    # display the progress of gradient descent
    cost = J2(theta_0_gd,theta_1_gd)
    print("theta_0: % f, theta_1: % f, J(theta_0,theta_1)= %f" %(theta_0_gd, theta_1_gd, cost))
    
    # find new parameters and weights
    theta_0_gd_n = theta_0_gd - alpha * 4*theta_0_gd
    theta_1_gd_n = theta_1_gd - alpha * 2* theta_1_gd
    
    theta_0_gd = theta_0_gd_n
    theta_1_gd = theta_1_gd_n
    
print("theta_0: % f, theta_1: % f, J(theta_0,theta_1)= %f" %(theta_0_gd, theta_1_gd, J2(theta_0_gd,theta_1_gd)))
```

## Exercises: Gradient Descent for Simple Linear Regression


Before you start coding, let'first define the simple linear regression model (equations).

A simple linear regression uses a linear function to describe the relationship between two variables. You can predict your target value y based on one feature x.

$$ y^{(i)} \approx \hat{y} =  h_\theta(\mathbf{x})$$

- Hypothesis for a simple linear regression:
$$
h_\theta(x^{(i)})= \theta_0 + \theta_1x_1^{(i)}
$$
- Cost function for a simple linear regression:
$$
J(\theta) = \frac{1}{2} \sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})^2 = \frac{1}{2} \sum_{i=1}^{m}(\theta_0 + \theta_1x_1^{(i)}-y^{(i)})^2
$$
- Gradient, partial derivatives of cost function:
$$
\frac{\partial}{\partial \theta_j} J(\theta) = \frac{\partial}{\partial \theta_j} \frac{1}{2} \sum_{i=1}^{m}(h_\theta(x^{(i)})-y^{(i)})^2
$$
